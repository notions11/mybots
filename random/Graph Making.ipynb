```python
import os
import matplotlib.pyplot as plt

path = "fitness/"
plot_dict = dict()

for file in os.listdir(path):
    plot1 = []
    with open(path + file) as f:
        plot1 = [float(line.split(',')[1].strip()) for line in f.readlines()]
        plot_dict[file.split('.')[0][11:]] = plot1
```

```python
ctr = 1

plt.figure(figsize=(10,6))

final_fitness = 0

for k in plot_dict.keys():
    print(k)
    plt.plot(list(range(1, len(plot_dict[k])+1)), plot_dict[k], label = "Run " + str(ctr), alpha=0.5)
    final_fitness += plot_dict[k][-1]
    ctr += 1
    
plt.gca().invert_yaxis()
plt.xlabel("Generation")
plt.ylabel("Fitness Score (Lower # Better)")
plt.legend(prop={'size': 5})
plt.show()

print(final_fitness/20.0)
```

```python
titles = ["Change to a \nSensor/Non-Sensor", "Change Size", "Move", "Add or Remove", "Change Weights"]
values = [64, 19, 10, 16, 61]
```

```python
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(titles, values)
 
plt.xlabel("Evolution Type")
plt.ylabel("Aggregate Count in final robots")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(values, labels=titles)
```
